```python
import pandas as pd
from datasets import Dataset
from sklearn.model_selection import train_test_split

# Gi·∫£ s·ª≠ df l√† DataFrame ban ƒë·∫ßu c·ªßa b·∫°n
# df = pd.read_excel('/mnt/data/your_data_file.xlsx') # Uncomment this if you need to load data from an Excel file

# Chia d·ªØ li·ªáu th√†nh t·∫≠p hu·∫•n luy·ªán v√† t·∫≠p t·∫°m th·ªùi (80% train, 20% temp)
train_df, temp_df = train_test_split(df, test_size=0.2, random_state=42)

# Chia t·∫≠p t·∫°m th·ªùi th√†nh t·∫≠p ki·ªÉm tra (10%) v√† t·∫≠p validation (10%)
val_df, test_2_df = train_test_split(temp_df, test_size=0.5, random_state=42)

# Chuy·ªÉn ƒë·ªïi DataFrame th√†nh Dataset c·ªßa th∆∞ vi·ªán datasets
train_dataset = Dataset.from_pandas(train_df)
test_dataset = Dataset.from_pandas(val_df)  # val_df b√¢y gi·ªù l√† test_dataset
test_2_dataset = Dataset.from_pandas(test_2_df)  # test_2_df b√¢y gi·ªù l√† test_dataset th·ª© hai

# L∆∞u test_2_dataset v√†o t·ªáp Excel tr√™n Kaggle/working
test_2_df.to_excel('/kaggle/working/test_2_dataset.xlsx', index=False)

print(train_dataset)
print(test_dataset)
print(test_2_dataset)
```

```python
from transformers import AutoTokenizer

# Kh·ªüi t·∫°o tokenizer
tokenizer = AutoTokenizer.from_pretrained("vinai/phobert-base-v2")
# tokenizer = AutoTokenizer.from_pretrained("5CD-AI/visobert-14gb-corpus")

max_length = 256

def tokenize_function(examples):
    texts = [str(text) if text is not None else "" for text in examples['content']]
    print("Processing batch size:", len(texts))  # Th√™m d√≤ng n√†y ƒë·ªÉ log k√≠ch th∆∞·ªõc batch
    tokenized_outputs = tokenizer(texts, padding=False, truncation=True, max_length=max_length)
    return tokenized_outputs

# Gi·∫£ s·ª≠ b·∫°n ƒë√£ c√≥ 'train_dataset' v√† 'test_dataset' l√† c√°c ƒë·ªëi t∆∞·ª£ng Dataset
train_dataset = train_dataset.map(tokenize_function, batched=True, batch_size=1000)
test_dataset = test_dataset.map(tokenize_function, batched=True, batch_size=1000)
```

```python
import os
from transformers import AutoModelForSequenceClassification, Trainer, TrainingArguments, DataCollatorWithPadding, AutoTokenizer, TrainerCallback
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, matthews_corrcoef, roc_auc_score

# Thi·∫øt l·∫≠p bi·∫øn m√¥i tr∆∞·ªùng cho API key c·ªßa WandB
os.environ["WANDB_API_KEY"] = "8acff34accfdb332022494ebe36901cb3ea8b15d"  # Thay th·∫ø b·∫±ng API key c·ªßa b·∫°n
os.environ["CUDA_LAUNCH_BLOCKING"] = "1"

# Kh·ªüi t·∫°o model v√† tokenizer
model = AutoModelForSequenceClassification.from_pretrained("vinai/phobert-base-v2", num_labels=2)
tokenizer = AutoTokenizer.from_pretrained("vinai/phobert-base-v2")
# model = AutoModelForSequenceClassification.from_pretrained("5CD-AI/visobert-14gb-corpus", num_labels=2)
# tokenizer = AutoTokenizer.from_pretrained("5CD-AI/visobert-14gb-corpus")

def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    pred_probs = pred.predictions[:, 1]  # L·∫•y x√°c su·∫•t cho l·ªõp d∆∞∆°ng t√≠nh (label=1)
    
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='binary')
    acc = accuracy_score(labels, preds)
    mcc = matthews_corrcoef(labels, preds)
    roc_auc = roc_auc_score(labels, pred_probs)
    
    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall,
        'mcc': mcc,
        'roc_auc': roc_auc
    }

# Custom Callback ƒë·ªÉ in ra c√°c th√¥ng s·ªë ƒë√°nh gi√°
class CustomCallback(TrainerCallback):
    def on_evaluate(self, args, state, control, metrics=None, **kwargs):
        if metrics is not None:
            print(f"Step {state.global_step}: {metrics}")

training_args = TrainingArguments(
    output_dir="/kaggle/working",  # ƒê·∫£m b·∫£o s·ª≠ d·ª•ng th∆∞ m·ª•c n√†y ƒë·ªÉ l∆∞u tr·ªØ t·∫°m th·ªùi
    num_train_epochs=6,
    per_device_train_batch_size=32,
    per_device_eval_batch_size=32,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=500,  # Th√™m logging_steps ƒë·ªÉ ƒë·ªãnh k·ª≥ ghi l·∫°i c√°c th√¥ng s·ªë ƒë√°nh gi√°
    save_steps=500,
    save_total_limit=2,
    evaluation_strategy="steps",  # Th√™m evaluation strategy ƒë·ªÉ ƒë√°nh gi√° theo t·ª´ng b∆∞·ªõc
    eval_steps=500,  # ƒê√°nh gi√° sau m·ªói 500 b∆∞·ªõc
    load_best_model_at_end=True,  # T·∫£i m√¥ h√¨nh t·ªët nh·∫•t sau khi hu·∫•n luy·ªán
    metric_for_best_model="accuracy",  # S·ª≠ d·ª•ng ƒë·ªô ch√≠nh x√°c ƒë·ªÉ ch·ªçn m√¥ h√¨nh t·ªët nh·∫•t
    greater_is_better=True,  # ƒê·ªô ch√≠nh x√°c cao h∆°n l√† t·ªët h∆°n
)

data_collator = DataCollatorWithPadding(tokenizer=tokenizer, return_tensors="pt")

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
    callbacks=[CustomCallback()]  # Th√™m CustomCallback v√†o Trainer
)

trainer.train()

# L∆∞u model v√† tokenizer v√†o th∆∞ m·ª•c l√†m vi·ªác
# model.save_pretrained("/kaggle/working/phobert_spam_model_stn")
# tokenizer.save_pretrained("/kaggle/working/phobert_spam_tokenize_stn")
# L∆∞u model v√† tokenizer v√†o th∆∞ m·ª•c l√†m vi·ªác
model.save_pretrained("/kaggle/working/phobert_model_best2_cut")
tokenizer.save_pretrained("/kaggle/working/phobert_tokenize_best2_cut")
```
