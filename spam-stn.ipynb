```python
import pandas as pd
from datasets import Dataset
from sklearn.model_selection import train_test_split

# Giả sử df là DataFrame ban đầu của bạn
# df = pd.read_excel('/mnt/data/your_data_file.xlsx') # Uncomment this if you need to load data from an Excel file

# Chia dữ liệu thành tập huấn luyện và tập tạm thời (80% train, 20% temp)
train_df, temp_df = train_test_split(df, test_size=0.2, random_state=42)

# Chia tập tạm thời thành tập kiểm tra (10%) và tập validation (10%)
val_df, test_2_df = train_test_split(temp_df, test_size=0.5, random_state=42)

# Chuyển đổi DataFrame thành Dataset của thư viện datasets
train_dataset = Dataset.from_pandas(train_df)
test_dataset = Dataset.from_pandas(val_df)  # val_df bây giờ là test_dataset
test_2_dataset = Dataset.from_pandas(test_2_df)  # test_2_df bây giờ là test_dataset thứ hai

# Lưu test_2_dataset vào tệp Excel trên Kaggle/working
test_2_df.to_excel('/kaggle/working/test_2_dataset.xlsx', index=False)

print(train_dataset)
print(test_dataset)
print(test_2_dataset)
```

```python
from transformers import AutoTokenizer

# Khởi tạo tokenizer
tokenizer = AutoTokenizer.from_pretrained("vinai/phobert-base-v2")
# tokenizer = AutoTokenizer.from_pretrained("5CD-AI/visobert-14gb-corpus")

max_length = 256

def tokenize_function(examples):
    texts = [str(text) if text is not None else "" for text in examples['content']]
    print("Processing batch size:", len(texts))  # Thêm dòng này để log kích thước batch
    tokenized_outputs = tokenizer(texts, padding=False, truncation=True, max_length=max_length)
    return tokenized_outputs

# Giả sử bạn đã có 'train_dataset' và 'test_dataset' là các đối tượng Dataset
train_dataset = train_dataset.map(tokenize_function, batched=True, batch_size=1000)
test_dataset = test_dataset.map(tokenize_function, batched=True, batch_size=1000)
```

```python
import os
from transformers import AutoModelForSequenceClassification, Trainer, TrainingArguments, DataCollatorWithPadding, AutoTokenizer, TrainerCallback
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, matthews_corrcoef, roc_auc_score

# Thiết lập biến môi trường cho API key của WandB
os.environ["WANDB_API_KEY"] = "8acff34accfdb332022494ebe36901cb3ea8b15d"  # Thay thế bằng API key của bạn
os.environ["CUDA_LAUNCH_BLOCKING"] = "1"

# Khởi tạo model và tokenizer
model = AutoModelForSequenceClassification.from_pretrained("vinai/phobert-base-v2", num_labels=2)
tokenizer = AutoTokenizer.from_pretrained("vinai/phobert-base-v2")
# model = AutoModelForSequenceClassification.from_pretrained("5CD-AI/visobert-14gb-corpus", num_labels=2)
# tokenizer = AutoTokenizer.from_pretrained("5CD-AI/visobert-14gb-corpus")

def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    pred_probs = pred.predictions[:, 1]  # Lấy xác suất cho lớp dương tính (label=1)
    
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='binary')
    acc = accuracy_score(labels, preds)
    mcc = matthews_corrcoef(labels, preds)
    roc_auc = roc_auc_score(labels, pred_probs)
    
    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall,
        'mcc': mcc,
        'roc_auc': roc_auc
    }

# Custom Callback để in ra các thông số đánh giá
class CustomCallback(TrainerCallback):
    def on_evaluate(self, args, state, control, metrics=None, **kwargs):
        if metrics is not None:
            print(f"Step {state.global_step}: {metrics}")

training_args = TrainingArguments(
    output_dir="/kaggle/working",  # Đảm bảo sử dụng thư mục này để lưu trữ tạm thời
    num_train_epochs=6,
    per_device_train_batch_size=32,
    per_device_eval_batch_size=32,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=500,  # Thêm logging_steps để định kỳ ghi lại các thông số đánh giá
    save_steps=500,
    save_total_limit=2,
    evaluation_strategy="steps",  # Thêm evaluation strategy để đánh giá theo từng bước
    eval_steps=500,  # Đánh giá sau mỗi 500 bước
    load_best_model_at_end=True,  # Tải mô hình tốt nhất sau khi huấn luyện
    metric_for_best_model="accuracy",  # Sử dụng độ chính xác để chọn mô hình tốt nhất
    greater_is_better=True,  # Độ chính xác cao hơn là tốt hơn
)

data_collator = DataCollatorWithPadding(tokenizer=tokenizer, return_tensors="pt")

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
    callbacks=[CustomCallback()]  # Thêm CustomCallback vào Trainer
)

trainer.train()

# Lưu model và tokenizer vào thư mục làm việc
# model.save_pretrained("/kaggle/working/phobert_spam_model_stn")
# tokenizer.save_pretrained("/kaggle/working/phobert_spam_tokenize_stn")
# Lưu model và tokenizer vào thư mục làm việc
model.save_pretrained("/kaggle/working/phobert_model_best2_cut")
tokenizer.save_pretrained("/kaggle/working/phobert_tokenize_best2_cut")
```
